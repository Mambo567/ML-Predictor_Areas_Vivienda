#### Se procede a utilizar las columnas que nos interesan.
De acuerdo con las conversaciones con el cliente, las columnas que nos interesan estan agrupadas asi:
Variables X (Variables que se tendrán al momento de hacer predicciones):
- El total de del área utilizable de la casa en metros cuadrados (m2).
- Número de habitaciones (Dormitorios)
- Número de Baños
- Número de aseos (Aquellos baños sin ducha)
- Si se tiene recibidor o no (Booleano)
- Si se tiene lavadero o no (Booleano)  

Las variables Y son las siguientes:
- Tamaño del área utilizable de las habitaciones en metros cuadrados (m2)
- Tamaño del área utilizable de la cocina en metros cuadrados (m2)
- Tamaño del área utilizable del salón-comedor en metros cuadrados (m2)
- Tamaño del área utilizable de los baños en metros cuadrados (m2)

```python
# Se dejan las columnas que nos interesan para el modelo. Nos interesan las de área, los perímetros no nos interesan de momento.
raw_data_areas = raw_data.loc[:,["total_m2","n_rooms","n_bathrooms","n_toilets","livingroom_m2","kitchen_m2","bathroom1_m2","bathroom2_m2","bathroom3_m2","toilet_m2","bdroom1_m2","bdroom2_m2","bdroom3_m2","bdroom4_m2","entrance_m2","corridor_m2","laundry_m2"]]
raw_data_areas
```

```python
# Cómo primer paso, vamos a validar que la sumatoria de las áreas de igual al total. Se procede a crear una columna adicional de total.
raw_data_areas.loc[:,"total_area"] = raw_data_areas.iloc[:,4:].sum(axis=1)
raw_data_areas
```

```python
# Validamos que el valor calculado sea igual que el valor proporcionado.
raw_data_areas.loc[:,"same_size"] = raw_data_areas.loc[:,"total_m2"].round() == raw_data_areas.loc[:,"total_area"].round()
```
